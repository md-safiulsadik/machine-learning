### 2 Creating a DataFrame from a dictionary

```python
# list used for this example
towns = ['Saptibari', 'Aditmari', 'Misson mor', 'Gosala bazar']
shops = [432, 546, 756, 3459]
```

```python
# Storing lists within a dictionary
dict_towns = {'Towns':towns, 'Shops':shops}
```

```python
# Creating the dataframe
df_towns = pd.DataFrame(dict_towns)
```

```python
# Showing the DataFrame
df_towns
```
